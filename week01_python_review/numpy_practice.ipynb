## NumPy 总结


- `np.array()` 用于创建向量，矩阵， 整体运算（区别List)；
- `shape` 表示数组形状(几行几列）,`ndim` 表示数组维度；
- `reshape()` 用于改变数组形状， -1自适应；
- 索引和切片可以取出数组中的部分数据；
- 广播机制可以让不同形状的数组进行运算, 低维扩高维；
- `@` 可以进行矩阵乘法；
- `mean()`、`std()`、`max()`、`min()` 是常用统计函数。


```python
import numpy as np

print("NumPy version:", np.__version__)
```

```python
# 从 Python 列表创建 NumPy 数组
a = np.array([1, 2, 3, 4, 5])

print(a)
print(type(a))
```

```python
zeros = np.zeros((3, 4))
ones = np.ones((2, 3))
random_array = np.random.randn(4, 5)

print("zeros:")
print(zeros)

print("ones:")
print(ones)

print("random_array:")
print(random_array)
```

```python
x = np.array([[1, 2, 3], [4, 5, 6]])

print("x:")
print(x)

print("shape:", x.shape)
print("dtype:", x.dtype)
print("维度数量:", x.ndim)
```

```python
x = np.arange(12)

print("原始 x:")
print(x)
print("原始 shape:", x.shape)

x_reshaped = x.reshape(3, 4)

print("reshape 之后:")
print(x_reshaped)
print("新的 shape:", x_reshaped.shape)
```

```python
x = np.arange(12).reshape(3, 4)

print("x:")
print(x)

print("第 0 行:")
print(x[0])

print("第 1 列:")
print(x[:, 1])

print("前两行，前两列:")
print(x[:2, :2])
```

```python
x = np.array([[1, 2, 3], [4, 5, 6]])

print("原始 x:")
print(x)

print("x + 10:")
print(x + 10)

y = np.array([10, 20, 30])

print("x + y:")
print(x + y)
```

```python
A = np.random.randn(3, 4)
B = np.random.randn(4, 2)

C = A @ B

print("A shape:", A.shape)
print("B shape:", B.shape)
print("C shape:", C.shape)

print("C:")
print(C)
```

```python
data = np.random.randn(100)

print("均值 mean:", data.mean())
print("标准差 std:", data.std())
print("最大值 max:", data.max())
print("最小值 min:", data.min())
print("中位数 median:", np.median(data))
```

```python
scores = np.array([60, 70, 80, 90, 100])

normalized_scores = (scores - scores.min()) / (scores.max() - scores.min())

print("原始分数:")
print(scores)

print("归一化之后:")
print(normalized_scores)
```

```python
epochs = np.arange(1, 11)
loss = np.exp(-epochs / 3) + np.random.randn(10) * 0.02

print("epochs:")
print(epochs)

print("loss:")
print(loss)
```
