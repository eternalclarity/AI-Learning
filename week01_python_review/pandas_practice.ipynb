## Pandas 总结

- DataFrame 可以理解成一张表；
- 每一行通常表示一条样本；
- 每一列通常表示一个特征；
- `head()` 可以查看前几行；
- `describe()` 可以查看统计信息；
- `df[df["score"] > 80]` 可以筛选数据；
- `sort_values()` 可以排序；
- `groupby()` 可以分组统计；
- `to_csv()` 和 `read_csv()` 可以保存和读取 CSV 文件；
- `isnull()` 和 `fillna()` 可以处理缺失值。


```python
import pandas as pd

print("Pandas version:", pd.__version__)
```

```python
data = {
    "name": ["Tom", "Jerry", "Alice", "Bob", "Mike", "Lucy"],
    "gender": ["male", "male", "female", "male", "male", "female"],
    "score": [85, 90, 78, 92, 66, 88],
    "study_hours": [5, 6, 3, 7, 2, 6],
}

df = pd.DataFrame(data)

df
```

```python
print("前 5 行:")
display(df.head())

print("数据形状:")
print(df.shape)

print("列名:")
print(df.columns)

print("数据类型:")
print(df.dtypes)
```

```python
df.describe()
```

```python
names = df["name"]
scores = df["score"]

print("姓名:")
print(names)

print("分数:")
print(scores)
```

```python
high_score_students = df[df["score"] >= 85]

high_score_students
```

```python
result = df[(df["score"] >= 85) & (df["study_hours"] >= 5)]

result
```

```python
df_sorted = df.sort_values(by="score", ascending=False)

df_sorted
```

```python
df["passed"] = df["score"] >= 60

df
```

```python
gender_score_mean = df.groupby("gender")["score"].mean()

gender_score_mean
```

```python
csv_path = "week1_python_review/student_scores.csv"

df.to_csv(csv_path, index=False, encoding="utf-8-sig")

print("CSV 文件已保存到:", csv_path)
```

```python
df_loaded = pd.read_csv("week1_python_review/student_scores.csv")

df_loaded
```

```python
data_with_missing = {
    "name": ["Tom", "Jerry", "Alice", "Bob"],
    "score": [85, None, 78, 92],
    "study_hours": [5, 6, None, 7],
}

df_missing = pd.DataFrame(data_with_missing)

df_missing
```

```python
df_missing.isnull()
```

```python
df_filled = df_missing.copy()

df_filled["score"] = df_filled["score"].fillna(df_filled["score"].mean())
df_filled["study_hours"] = df_filled["study_hours"].fillna(df_filled["study_hours"].mean())

df_filled
```
