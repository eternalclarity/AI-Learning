# Matplotlib Practice

本 Notebook 用于练习 Matplotlib 的基础绘图。

本次练习内容包括：

1. 折线图；
2. loss 曲线；
3. accuracy 曲线；
4. 柱状图；
5. 散点图；
6. 直方图；
7. 保存图片。

后面训练深度学习模型时，经常需要画 loss 和 accuracy 曲线，所以本节非常重要。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Line Plot")
plt.show()
```

```python
epochs = np.arange(1, 11)
loss = [1.20, 1.05, 0.92, 0.80, 0.68, 0.57, 0.49, 0.42, 0.36, 0.31]

plt.plot(epochs, loss, marker="o")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss Curve")
plt.grid(True)
plt.show()
```

```python
accuracy = [0.52, 0.58, 0.64, 0.69, 0.74, 0.78, 0.82, 0.85, 0.87, 0.89]

plt.plot(epochs, accuracy, marker="o")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training Accuracy Curve")
plt.grid(True)
plt.show()
```

```python
plt.plot(epochs, loss, marker="o", label="Loss")
plt.plot(epochs, accuracy, marker="s", label="Accuracy")

plt.xlabel("Epoch")
plt.title("Loss and Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

```python
models = ["LogReg", "SVM", "RandomForest", "MLP"]
f1_scores = [0.78, 0.82, 0.86, 0.84]

plt.bar(models, f1_scores)
plt.xlabel("Model")
plt.ylabel("F1 Score")
plt.title("Model Comparison")
plt.ylim(0, 1)
plt.show()
```

```python
study_hours = [2, 3, 4, 5, 6, 7, 8]
scores = [60, 65, 70, 78, 85, 90, 95]

plt.scatter(study_hours, scores)
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.title("Study Hours vs Score")
plt.grid(True)
plt.show()
```

```python
random_scores = np.random.normal(loc=75, scale=10, size=100)

plt.hist(random_scores, bins=10)
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.title("Score Distribution")
plt.show()
```

```python
epochs = np.arange(1, 11)
loss = [1.20, 1.05, 0.92, 0.80, 0.68, 0.57, 0.49, 0.42, 0.36, 0.31]

plt.plot(epochs, loss, marker="o")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss Curve")
plt.grid(True)

save_path = "week1_python_review/loss_curve.png"
plt.savefig(save_path, dpi=300, bbox_inches="tight")

plt.show()

print("图片已保存到:", save_path)
```

## Matplotlib 总结

本节练习了 Matplotlib 的基础绘图。

需要重点掌握：

- `plt.plot()` 用于画折线图；
- `plt.bar()` 用于画柱状图；
- `plt.scatter()` 用于画散点图；
- `plt.hist()` 用于画直方图；
- `plt.xlabel()` 和 `plt.ylabel()` 用于设置坐标轴名称；
- `plt.title()` 用于设置标题；
- `plt.legend()` 用于显示图例；
- `plt.grid()` 用于显示网格；
- `plt.savefig()` 用于保存图片。

后面训练模型时，loss 曲线和 accuracy 曲线是最常见的实验可视化方式。
