```python
def call(name: str, age: int) -> bool:
    print(f"hello {name}, are your age lower 18?", end="")
    if (age < 18):
        return True
    else:
        return False


name = "Tom"
age = 17
print(call(name, age))
```

```python
def reverse_words(s: str) -> str:
    # 字符串以空格分割成一个个单词，并组成列表
    words = s.split(' ')

    # 反转列表
    words = words[-1::-1]

    # 重新组合成字符串
    return ' '.join(words)


s = "I love to see you"
output = reverse_words(s)
print(output)
```
