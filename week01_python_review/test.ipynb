```python
def reverse_words(s: str = "I love to see you") -> str:
    """
    逆转字符串
    """
    
    # 字符串以空格分割成一个个单词，并组成列表
    words = s.split()

    # 反转列表
    words = words[-1::-1]

    # 重新组合成字符串
    return ' '.join(words)
print(reverse_words())
```

```python
def collect_ch(s: str = "I love to see you") -> list:
    """
    统计出现的字符
    """
    
    set_ch = set(s)
    
    l = []
    for c in set_ch:
        l.append(c)
    
    return l
print(collect_ch())
```

```python
def count_ch(s: str = "I love to see you") -> dict:
    """
    统计字符出现的词书
    """
    mp = {}
    
    for c in s:
        mp[c] = mp.get(c, 0) + 1
    
    return mp
print(collect_ch())
```

```python
def logging(log_file: str = "week1_python_review/scripts/train.log") -> None:
    import logging
    from pathlib import Path

    log_path = Path(log_file)
    log_path.parent.mkdir(parents=True, exist_ok=True)
    
    logging.basicConfig(
        level=logging.INFO,
        format="%(asctime)s [%(levelname)s] %(message)s",
        handlers=[
            logging.StreamHandler(),
            logging.FileHandler(log_path, encoding="utf-8"),
        ],
    )

    logging.info("程序开始运行")
    logging.warning("这是警告")
    logging.error("这是错误")
```
